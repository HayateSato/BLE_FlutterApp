```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# checking the name of available datasets
sns.get_dataset_names()
```

```python
# importing dataset
data = sns.load_dataset("car_crashes")
data.info()
```

```python
# showing the relationship between each cols
sns.pairplot(data, height=1.25)
plt.show()
```

```python
# showing the correlation
data_ = data.drop(["abbrev"], axis=1)
display(data_.corr())

# visualising the correrlation
sns.heatmap(data_.corr(), annot=True, cmap="Blues")
plt.show()
```
